#### Q4. Now the scraping. Please scrape the information from the webpage, with the following requirements: 

1. you are expected to scrape a set of information from the web, i.e., elements in the tables, elements in several tags, several files, or any other information that fit your purposes (5 marks);

```python
#find elements from soup
import requests
from bs4 import BeautifulSoup
import pandas as pd
url='https://weather.mipang.com/tianqi-51031/15tian.html'
re = requests.get(url)
re.encoding = 'utf-8'
soup = BeautifulSoup(re.text,'html.parser')
weather_container = soup.find('div', attrs={'class': 'clearfix body'})
```

```python
day = [d.text for d in weather_container.find_all(class_='tt t2')]
temp = [temp.text for temp in weather_container.find_all(class_='tt t5')]
img = weather_container.find_all('img')
```

```python
print(day)
print(temp)
print(img)
```

2. you are expected to use at least one of the control flow statements to fine tune the code (5 marks);

```python
img_list=list()
for txt in img:
    print(txt.get('src'))
    img_list.append(txt.get('src'))
```

3. You are expected to store the information in a structured format, and output the file in CSV format (2 marks). Attention: please use “utf-8” as the encoding. 

```python
df = {
    'day':day,
    'temp':temp,
    'img':img_list}
```

```python
pd_weather = pd.DataFrame.from_dict(df, orient ='index').transpose()
print(pd_weather.info())
```

```python
pd_weather
```

```python
pd_weather.to_csv('pd_weather.csv')
```

4. Bonus (5 marks, optional) is offered for scraping multiple pages, using pre-defined rules. 

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
url='https://weather.mipang.com/changsha/15tian.html'
re = requests.get(url)
re.encoding = 'utf-8'
soup = BeautifulSoup(re.text,'html.parser')
weather_container = soup.find('div', attrs={'class': 'clearfix body'})
day1 = [d.text for d in weather_container.find_all(class_='tt t2')]
temp1 = [temp.text for temp in weather_container.find_all(class_='tt t5')]
des1 = [des.text for des in weather_container.find_all(class_='tt t4')]
```

```python
cs = pd.DataFrame({
    'day':day1,
    'temp':temp1,
    'desc':des1})
cs_weather = pd.DataFrame.from_dict(cs)
print(cs_weather) #weather of 30 days in ChangSha
cs_weather.to_csv('cs_weather.csv',encoding='utf_8_sig')
```
